```python
def df_x(x):
    df = 6*x
    return df
```

```python
def f(x, y):
    f1 = 3*x**2 + 5*y**2
    return f1
def df(x, y):
    df1 = 6*x
    df2 = 10*y
    return df1, df2
```

```python
res = []
x0 = 5
y0 = 10
while True:
    res.append([x0, y0, f(x0, y0)])
    df1, df2 = df(x0, y0)
    x1 = x0 - l*df1
    y1 = y0-l*df2
    if np.linalg.norm(np.array([x1-x0])-np.array([y1-y0])) <= 0.00000001:
        break
    x0 = x1
    y0 = y1
print(res)
```

```python
res = np.array(res)
res
```

```python
x_values = res[:, 0]
y_values = res[:, 1]
z_values = res[:, 2]

#     
x = np.linspace(-6, 6, 100)
y = np.linspace(-6, 6, 100)
X, Y = np.meshgrid(x, y)
Z = 3*X + 5*Y

plt.figure(figsize=(10, 8))

#   
plt.contour(X, Y, Z, levels=50, cmap='Blues')

#    
plt.scatter(x_values, y_values, c='red', label='Gradient Descent Path')
plt.plot(x_values, y_values, c='red')

plt.title('Gradient Descent Visualization')
plt.xlabel('X')
plt.ylabel('Y')
plt.colorbar(label='Function Value')
plt.legend()
plt.show()
```
